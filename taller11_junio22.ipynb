Gráfico cuantil-cuantil
=======================

```python
# Importar librerías

import numpy as np
from matplotlib import pyplot as plt
```

Simular datos de acuerdo a dos distribuciones normales: $N(0,1)$ y $N(-2,5)$

```python
x=np.random.normal(0,1,500) # usar como referencia
y=np.random.normal(-2,5,100)   # simulación de datos experimentales
```

```python
q=np.linspace(0,1,len(y))
q = q[1:-1]
xq=np.quantile(x,q)
yq=np.quantile(y,q)
```

```python
plt.hist(x,bins=50)
for i in range(len(xq)):
    plt.axvline(x=xq[i],color="orange",alpha=0.15)
```

```python
plt.scatter(xq,yq)
plt.show()
```

```python
# O más rapidamente a través de sicpy
from scipy import stats

res = stats.probplot(y, plot=plt)
```

## Hace el gráfico:

1. cuantil-cuantil de los residuos.
2. graficar los residuos en el orden en que su respectiva muestra fue tomada.

```python
data=np.array([[575,542,530,539,570],
               [565,593,590,579,610],
               [600,651,610,637,629],
               [725,700,715,685,710]]) 
```

```python
residuales=np.zeros ((4,5)) 
residuales_list=[] 
for i in range (4): 
    for j in range(5): 
        y_bar=np.mean(data[i,:]) 
        residuales[i,j]=data[i,j]-y_bar 
        residuales_list.append(data[i,j]-y_bar) 
```

```python
y=residuales_list
x=np.random.normal(0,1,1000) # usar como referencia

q=np.linspace(0,1,len(y))
q=q[1:-1]
xq=np.quantile(x,q)
yq=np.quantile(y,q)
plt.scatter(xq,yq)
```

```python
res = stats.probplot(y, plot=plt)
```
